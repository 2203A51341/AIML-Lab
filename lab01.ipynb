#Data Visualization using Matplotlib

```python
people=['a','b','c','d','e']
age=['42','36','21','18','4']
weight=['70','80','65','55','15']
height=['170','178','179','173','90']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.show()
```

#Bar Chart

```python
plt.figure(figsize=(5,4))
plt.title("People's weight in kilograms:",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people, height= weight, width=0.6, color='orange', edgecolor='k', alpha =0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight in kilograms:",fontsize=15)
plt.show()
```

#Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='pink',edgecolor='k',alpha=0.75,bins=5)
plt.xlabel("histogram of patient weight",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.xticks(fontsize=15,rotation=30)
plt.yticks(fontsize=15)
plt.show()
```
